## These are the plots comparing the results when the dispersion velocity is constant

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

# Velocity curve plot
ax[0].plot(inpc['RADII'],inpc['VROT'], color='black', linewidth=2, label='input')
ax[0].errorbar(a_vrot["Radius"],a_vrot["Median"],capsize=5,fmt='o', markersize=7,yerr=[a_vrot["Error_Low"],a_vrot["Error_Up"]],label='fixed 180', color= col[0])
ax[0].errorbar(b_vrot["Radius"],b_vrot["Median"],capsize=5,fmt='s', markersize=7,yerr=[b_vrot["Error_Low"],b_vrot["Error_Up"]],label='fixed 360', color= col[1])
ax[0].errorbar(c_vrot["Radius"],c_vrot["Median"],capsize=5,fmt='o', markersize=7,yerr=[c_vrot["Error_Low"],c_vrot["Error_Up"]],label='semi free 180', color= col[2])
ax[0].errorbar(d_vrot["Radius"],d_vrot["Median"],capsize=5,fmt='s', markersize=7,yerr=[d_vrot["Error_Low"],d_vrot["Error_Up"]],label='semi free 360', color= col[3])
ax[0].errorbar(e_vrot["Radius"],e_vrot["Median"],capsize=5,fmt='o', markersize=7,yerr=[e_vrot["Error_Low"],e_vrot["Error_Up"]],label='free 180', color= col[4])
ax[0].errorbar(f_vrot["Radius"],f_vrot["Median"],capsize=5,fmt='s', markersize=7,yerr=[f_vrot["Error_Low"],f_vrot["Error_Up"]],label='free 360', color= col[5])
ax[0].set_xlabel('Radius (arcsec)')
ax[0].set_ylabel('Rotational velocity (km/s)')
ax[0].legend(fontsize = "x-large")

#Dispersion plot
ax[1].plot(inpc['RADII'],inpc['VDISP'], color='black', linewidth=2, label='input')
ax[1].errorbar(a_vdisp["Radius"],a_vdisp["Median"],capsize=5,fmt='o', markersize=7,yerr=[a_vdisp["Error_Low"],a_vdisp["Error_Up"]],label='fixed 180', color= col[0])
ax[1].errorbar(b_vdisp["Radius"],b_vdisp["Median"],capsize=5,fmt='s', markersize=7,yerr=[b_vdisp["Error_Low"],b_vdisp["Error_Up"]],label='fixed 360', color= col[1])
ax[1].errorbar(c_vdisp["Radius"],c_vdisp["Median"],capsize=5,fmt='o', markersize=7,yerr=[c_vdisp["Error_Low"],c_vdisp["Error_Up"]],label='semi free 180', color= col[2])
ax[1].errorbar(d_vdisp["Radius"],d_vdisp["Median"],capsize=5,fmt='s', markersize=7,yerr=[d_vdisp["Error_Low"],d_vdisp["Error_Up"]],label='semi free 360', color= col[3])
ax[1].errorbar(e_vdisp["Radius"],e_vdisp["Median"],capsize=5,fmt='o', markersize=7,yerr=[e_vdisp["Error_Low"],e_vdisp["Error_Up"]],label='free 180', color= col[4])
ax[1].errorbar(f_vdisp["Radius"],f_vdisp["Median"],capsize=5,fmt='s', markersize=7,yerr=[f_vdisp["Error_Low"],f_vdisp["Error_Up"]],label='free 360', color= col[5])
ax[1].set_ylim(0, 40)
ax[1].set_xlabel('Radius (arcsec)')
ax[1].set_ylabel('Dispersion velocity (km/s)')
ax[1].legend(fontsize = "x-large")

plt.tight_layout()
plt.show()
```
